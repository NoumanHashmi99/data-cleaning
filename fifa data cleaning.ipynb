```python
def convert_value(val):
    if 'M' in val:
        return float(val.replace('€', '').replace('M', '')) * 1000000
    elif 'K' in val:
        return float(val.replace('€', '').replace('K', '')) * 1000
    else:
        return float(val.replace('€', ''))  # Handle values without M or K
```

```python
fifa['Value'] = fifa['Value'].apply(convert_value)
```
